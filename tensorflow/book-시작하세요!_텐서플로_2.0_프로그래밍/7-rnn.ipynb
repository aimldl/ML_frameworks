## 곱셈 문제 데이터 생성

```python
import numpy as np

X = []
Y = []

for i in range(3000):
    lst = np.random.rand( 100 )
    idx = np.random.choice( 100, 2, replace=False )
    # Create a one-hot encoding vector
    zeros = np.zeros(100)
    zeros[idx] = 1
    X.append( np.array( list( zip( zeros, lst) )))
    Y.append( np.prod( lst[idx]) )
    
X[0], Y[0]
```
