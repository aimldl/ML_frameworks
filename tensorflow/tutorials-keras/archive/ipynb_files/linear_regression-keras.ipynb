````python
# Importing the necessary packages
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np
import matplotlib.pyplot as plt
````

````python
# Data Generation
x = np.array([
    [1, 2, 0], [5, 4, 3], [1, 2, -1], [3, 1, 0], [2, 4, 2],
    [4, 1, 2], [-1, 3, 2], [4, 3, 3], [0, 2, 6], [2, 2, 1],
    [1, -2, -2], [0, 1, 3], [1, 1, 3], [0, 1, 4], [2, 3, 3]
])

y = np.array([
    -4, 4, 6, 3, -4,
    9, -7, 5, 6, 0,
    4, 3, 5, 5, 1    
])

x.shape
````

````python
# Training data
x_train = x
y_train = y

# Test data
x_test_list = [
    [5, 5, 0], [2, 3, 1], [-1, 0, -1], [10, 5, 2], [4, -1, -2]
]
# y = 2*x0 + 3*x1 + 2*x2
y_test_list = [2*data[0] - 3*data[1] + 2*data[2] for data in x_test_list]

x_test = np.array( x_test_list )
y_test = np.array( y_test_list )
````

````python
# Data exploration
#   The training dataset is y = f(x) which is a 4D function.
#   Visualization is difficult, but we'll try to visually understand the function.
#
# Google search: matplotlib 3d plot numpy array
#   * Plotting a simple 3d numpy array using matplotlib
#       https://stackoverflow.com/questions/40351026/plotting-a-simple-3d-numpy-array-using-matplotlib

# Google search: matplotlib 3d plot
#   * matplotlib > mplot3d tutorial
#       https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(x[:, 0], x[:, 1], x[:, 2], 'ko') # ko = black dot
ax.set_title('x (in 3D) or x_train')
ax.set_xlabel('x[0]')
ax.set_ylabel('x[1]')
ax.set_zlabel('x[2]')
plt.show()
````

````python
plt.plot( y, 'ro' )  # ro = red dots
plt.title('y or y_train')
plt.ylabel('y')
plt.grid()

# matplotlib.pyplot.xticks > Examples > Disable xticks
#   https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.xticks.html
plt.xticks([])
````

````python
# Model Implementation

#   Losses
#     https://keras.io/api/losses/
#   MeanSquaredError class
#     https://keras.io/api/losses/regression_losses/#meansquarederror-class
#   Regression losses
#     https://keras.io/api/losses/regression_losses/#mean_squared_error-function

input_shape = (3,)
learning_rate = 1e-2
loss = 'mse'  # mean-squared-error
optimizer = SGD(learning_rate)
max_epoch = 1000

model = Sequential(
 [Dense(1, input_shape = input_shape, activation = 'linear')]
)
````

* `model.add()` Keras API를 사용해도 같은 결과를 얻을 수 있습니다.
```python
model = Sequential()
model.add(Dense(1, input_shape = input_shape, activation = 'linear'))
)
```
* 자세한 내용은 [케라스 Sequential 모델 시작하기](https://keras.io/ko/getting-started/sequential-model-guide/)를 참고하세요.

````python
model.compile(optimizer = optimizer, loss = loss)

# Double-check the model
model.summary()
````

````python
# Training
hist = model.fit(x_train, y_train, epochs=max_epoch)
````

````python
# Model evaluation
y_hat_test = model.predict(x_test)
````
